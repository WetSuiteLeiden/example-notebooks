## Purpose of this notebook

This is a list of legal and technical jargon (and abbreviations I keep seeing), 
only really added when understanding them was necessary for writing some of the notebooks here.

For a good part, these are **terms that popped repeatedly up while sifting through data, introductions and metadata**, and
- might be new to law students         
  - e.g. some NLP terms
  - say, when you're going through text, you may care which method you want between clustering an topic modeling and NER, without an in-depth course
- might be new to linguist programmers 
  - e.g. some _types_ of documents in a collection, and metadata it is often tagged with
  - say, when you're going through rechtspraak.nl data, you may care about _approximately_ what an Avv or a cassatie is _without_ detailed legal context from an in-depth course, or why some things have `-bes` on it
- might be new to to both
  - e.g. some XML parsing details
- might have have different meanings between these fields,
  - case in point: annotation

DUE WARNING: 
- this is intentionally simplified to the point of "I'll have to look up the details myself". 
- Also, it was written by a programmer, so particularly legal terms it may be oversimplified, incorrect, or _both_.

## A

**Akomo Ntoso**  - XML document format aimed at resonably encoding executive, legislative, and judiciary documents, with a controlled vocabulary
- [wikipedia](https://en.wikipedia.org/wiki/Akoma_Ntoso)

**AKN** - when not referring to the just-mentioned XML format, it often refers to the [Akoma Ntoso Naming Convention (AKN NC)](http://docs.oasis-open.org/legaldocml/akn-nc/v1.0/akn-nc-v1.0.html)
- that naming convention has a scope that narrows from left to right (vaguely URN-like), like `/akn/nl/act/gm0202/2023/Regeling3d27a1736497475a8534a65837a94519`
- where e.g. the second item is the country, the third is usually one of a handful of doctypes (and some optional things making life a little more complex)
- See also JOIN
- https://koop.gitlab.io/STOP/standaard/1.3.0/naamgevingsconventie.html

**AKN resolver** (arguably more fully an 'Akoma Ntoso IRI Resolver') - a service that resolves AKNs (and variants, see JOIN) to documents 
- ...within a certain system of documents; there does not seem to be a way to build a universal resolver
- a number of resolvers exist, and may be specific to a specific doctype (e.g. resolves only `/akn/nl/bill`) and/or maybe specific to the system/dataset they sit next to
- ...so you may need to query multiple to get an answer you consider complete, and you may need to find some sort of catalogue of options to find out where (see e.g. [this](https://repository.overheid.nl/frbr/cga/resolver/performers/1/json/servers.json))
- https://akresolver.cs.unibo.it/ https://www.overheid.nl/resolver/introductie

**Avv** (algemeen verbindende voorschriften) - roughly speaking, a rule that applies to everyone, and is binding, usually _based on_ a specific law (VERIFY)
- (more practical contrast somewhere below)
- [Aanwijzing 2.17 Begrip algemeen verbindend voorschrift](https://www.kcbr.nl/beleid-en-regelgeving-ontwikkelen/aanwijzingen-voor-de-regelgeving/hoofdstuk-2-algemene-onderwerpen-van-regelgeving/ss-23-algemeen-verbindende-voorschriften/aanwijzing-217-begrip-algemeen-verbindend-voorschrift) [Aanwijzingen voor de regelgeving - Aanwijzing 2.17. Begrip algemeen verbindend voorschrift](https://wetten.overheid.nl/BWBR0005730/2022-04-01/#Hoofdstuk2_Paragraaf2.3_Artikel2.17)

**Algemene wetten** - some laws are general in the sense that they have wider effect and reach than many others, and may be relatively central to a particular area of practice. Consider e.g. [those mentioned here (Aanwijzingen voor de regelgeving - Aanwijzing 2.46. Afwijking van algemene wetten)](https://wetten.overheid.nl/BWBR0005730/2022-04-01/#Hoofdstuk2_Paragraaf2.8_Artikel2.46)

**AMvB** (Algemene Maatregel van Bestuur)
- a few are independent (zelfstandige AMvB), but most are delegated legislation (gedelegeerde wetgeving), meaning they come from a specific law that says "that details have to be worked out separately"
- often ''relatively'' practical implementation details, like the method of distributing finances, how executive groups should be formed, etc.
- implicitly makes it easier to create the law itself, as well as to change these implementation details, as AMvBs are easier to add and change implementation details of <!--, e.g. to change drugs to the list the Opiumnwet applies to[https://zoek.officielebekendmakingen.nl/stb-2017-206.html]) than laws -->, e.g. often does not require [Staten-Generaal](#s), though should still involve [Raad van State](#r) (VERIFY)
- published in Staatsblad
<!--
- is a specific type of Koninklijk Besluit (KB) - and as such ''are'' contasigned by ministers, mostly for reasons of responsibility.
-->

**Annotation**
- **In the legal world**, annotation is a pragmatic human thing, a note to explain or illustrate meaning or help interpret a text, often of moderate sized chunks of text, sometimes even of the whole
- **In the computer/data world**, annotation is a simple mechanical thing, often demarking specific things like words and phrases, often to create wide training data to also recognize/mar/extract similar elsewhere, often each of a specific type, small indivisible things so that little fuzziness or interpretation is necessary for the machine learning that uses it (e.g. 'this is a common legal term', 'this is a reference to a law', 'this is the name of a well known case')

## B


**Beslissing** - decision, in a neutral sense and not necessarily relating to legalities at all.  Even in legal context, beslissing _may_ refer to nothing more than deciding to get coffee.

**Besluit** - a beslissing(decision) that has some sort of legal implication (rechtsgevolg), that _changes_ something.
- Often a yes-or-no sort of thing.  (VERIFY)
- pretty central in bestuursrecht (e.g. has a definition at [Artikel 1:3 Algemene wet bestuursrecht](https://wetten.overheid.nl/jci1.3:c:BWBR0005537&hoofdstuk=1&titeldeel=1.1&artikel=1:3&z=2018-01-01&g=2018-01-01))
  - may be mentioned as a besluit van algemene strekking (bas), to contrast it with one that is ''not'', i.e. a beschikking

**Beschikking** - besluit that applies to a specific case (rather than something that might apply in general)
- e.g. handling a specific person's case
- or handling a specific permit (in bestuursrecht, administrative law)
- or, in civiel recht (civil law), the final say in a procedure that starts with a verzoekschrift<!--
(petition?)  (instead of starting with a  dagvaarding(summons), which leads to a vonnis(judgment)).

**Consolidatie**  
- For context: 
  - a lot of legal text will be modified once or more. 
  - for laws this often comes in the form of _another_ law that says it augments the earlier version with a few details. The text of this agumenting law will be only those changes.
  - The _combination_ of these separate-but-together laws is what is really in effect, but hard to oversee, particularly for non-experts.
- Sometimes it us useful to present all these components as a set
- More usually, people care about what they have to adhere to today, so we prefer to see consolidations: a single single text with all updates included
  - Consolidations are considered informative, not legally binding, _roughly_ because it is considered redundant with the basic laws
  - but they _should_ be equivalent, so you can basically treat them as if they are
- When we look at resources that serve us text, it is useful to know whether it publishes the changes and/or the consolidated version

**Stb** - Staatsblad [Staatsblad van het Koninkrijk der Nederlanden](https://nl.wikipedia.org/wiki/Staatsblad_van_het_Koninkrijk_der_Nederlanden)

**STOP** - Standaard Officiële Publicaties [documentatie van versie 1.3.0 van de Standaard Officiële Publicaties ](https://koop.gitlab.io/STOP/standaard/1.3.0/index.html)

## T

**TOOI** - [Thesaurus en Ontologie Overheidsinformatie](https://tardis.overheid.nl/), from KOOP - an ontological model and theraurus
- e.g. the [waardelijsten](https://tardis.overheid.nl/waardelijsten.html) can be useful. Also used in STOP.

**TOP**-lijst (Thema-indeling voor Officiële Publicaties)
- a list/ontology of some broad themes to use for publications [Thema-indeling voor Officiële Publicaties (TOP-lijst)](https://standaarden.overheid.nl/tooi/waardelijsten/expression?lijst_uri=https%3A%2F%2Fidentifier.overheid.nl%2Ftooi%2Fset%2Fscw_toplijst%2F1)
- part of TOOI? (VERIFY)

**Trb** - **Tractatenblad** (... van het Koninkrijk der Nederlanden) - mainly treaties? [Tractatenblad](https://nl.wikipedia.org/wiki/Tractatenblad)
- considered part of the officiele bekendmakingen

**Topic modeling** - an NLP method that, given documents an little else, tries to estimate what _groups_ of repeated terms (~= topics) seem to reasonably divide and cluster documents within that given set

## U

**Uniforme Productnamenlijst** (**UPL**) - list/ontology (with URI as identifier) used to be clear about products and services aimed at people and businesses : from `aanschrijving` to `rijbewijs` to `gasaansluitingsverzoek` to `batterijen en accu's-inzamelverplichting`
- Currently approx 1300 entries
  - there are specific subsets for gemeenten, provincies, waterschappen, and Rijksoverheid
-  not sure who exactly uses them
- [description](https://standaarden.overheid.nl/upl) &middot; [expansion](https://standaarden.overheid.nl/upl/doc/Toelichting-UPL.pdf) &middot; [UPL-notities list with some notes](https://standaarden.overheid.nl/owms/oquery/)

## W

**Wob** - see Woo

**Woo** - **Wet open overheid** - one of a few laws ([BWBR0045754](https://wetten.overheid.nl/BWBR0045754)) that drives public (and digital) access to government documents
- replaces the earlier Wet openbaarheid van bestuur [(Wob)](https://nl.wikipedia.org/wiki/Wet_openbaarheid_van_bestuur)  (VERIFY)

**Woo-index** - a recently introduced website - the idea seems to eventially be a portal all (central and decentral) government organisations to whom the Woo applies
- https://www.rijksoverheid.nl/documenten/publicaties/2023/07/17/de-17-verplichte-categorieen-actief-openbaar-te-maken-informatie-volgens-de-woo

**Woo-verzoek** - request for information on the basis of [Wet open overheid (hoofdstuk 4, Openbaarmaking op verzoek)](https://wetten.overheid.nl/BWBR0045754/#Hoofdstuk4), to make public certain government information (e.g. if it isn't already) ([Vraag en antwoord Woo verzoeken](https://www.om.nl/onderwerpen/wet-open-overheid/veel-gestelde-vragen))
- towards Rijksoverheid, municipalities, provinces, waterschappen, Raad van State, de Nationale Ombudsman, de Raad voor de Rechtspraak, de Algemene Rekenkamer
- will be honored in the degree it is reasonable; there are a number of reasons parts (or even the whole) may not be

**Wsb** - **Waterschapsblad** - the Dutch [water board](https://en.wikipedia.org/wiki/Water_board) is historically a thing separate from provinces.[Waterschapsblad](https://www.koopoverheid.nl/officiele-overheidspublicaties/officiele-bekendmakingen/waterschapsblad)
- considered part of the officiele bekendmakingen

**WTI** - wetstechnische informatie (used in BWB, maybe more?)
- mostly
  - "things legally based in part on this"
  - "things that reference this"  (VERIFY) - including jci style links
  - changes (the start of the date this varies a little with content type, within 2002~2005)
- [Wetstechnische informatie regelingselement](https://www.overheid.nl/help/wet-en-regelgeving/wetstechnische-informatie-regelingselement)


## Z

**ZBO** - Zelfstandig BestuursOrgaan - organisations that perform a government task, interacting with but not part of a ministery [Zelfstandige bestuursorganen (zbo's)](https://www.rijksoverheid.nl/onderwerpen/rijksoverheid/zelfstandige-bestuursorganen).
- Some large, some tiny, from public service providers, driving tests, inspection bodies, some things related to health care and banking, trade registries, commissions, etc, see [a list here](https://organisaties.overheid.nl/Zelfstandige_bestuursorganen/). In context of legal text, we often see these when talking about rules that apply to them  (VERIFY)
