```python
import wetsuite.datasets
import wetsuite.helpers.lazy
import wetsuite.helpers.notebook
import wetsuite.extras.lawref
from importlib import reload
reload(wetsuite.datasets)

for _, xmlbytes in wetsuite.helpers.notebook.ProgressBar( wetsuite.datasets.load('cvdr-mostrecent-xml').data.items() ):
    # if b'extref' not in xmlbytes:
    #     continue
    tree = wetsuite.helpers.lazy.etree( xmlbytes )

    for extref in tree.iter(tag='extref'):
        #if extref.text is None:
        #    continue
        value = extref.attrib.get('doc', None)
        # that value is often an ID or URL
        #print(value)
        if value is None:
            continue
        value = value.strip()
        if value.startswith('http://wetten.overheid.nl/cgi-bin/deeplink/'):
            # will look like
            #   http://wetten.overheid.nl/cgi-bin/deeplink/law1/title=Burgerlijk%20Wetboek%20Boek%201
            #   http://wetten.overheid.nl/cgi-bin/deeplink/law1/bwbid=BWBR0005537/article=1:2
            # Those are not query parameters, and I've not found the standard this might be following, so some estimation is involved here
            print( value )
            print( wetsuite.extras.lawref.resolve_deeplink_bwbid( value ) )
```
