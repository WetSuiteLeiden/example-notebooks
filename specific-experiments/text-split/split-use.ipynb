```python
bwb_html = wetsuite.datasets.load( 'cvdr-mostrecent-html' )
```

```python
for id, docbytes in bwb_html.data.items():
    print( id )
    splitters = wetsuite.helpers.split.decide( docbytes )
    if len(splitters)==0:
        print( "DUNNO %s"%id )    
    for _,_,text in splitters[0][1].fragments():
        print( text )
        print( '--------------------' )

    break
```
