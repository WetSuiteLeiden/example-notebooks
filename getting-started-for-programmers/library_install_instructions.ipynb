# Installing wetsuite

Wetsuite as a project consists of a few distinct things, such as 
- [the website](https://www.wetsuite.nl/) and its guidance,
- some [datasets, as introduced here](https://github.com/WetSuiteLeiden/example-notebooks/tree/main/datasets-use)
- notebooks that are specific [introductions](https://github.com/WetSuiteLeiden/example-notebooks/tree/main/research-methods) or [experiments](https://github.com/WetSuiteLeiden/example-notebooks/tree/main/specific-experiments),
- our own [library of helpers](https://github.com/WetSuiteLeiden/wetsuite-core) that most of those notebooks rely on.

This is just about the last one or two:
installation of that library, often so that you can run those notebooks.

...because those notebooks rely on our own library (and a dozen more libraries that ours relies on).

**The short version** is that :

```python
!pip install -U wetsuite
```

...should install wetsuite, and all the things it depends on.

## Notes on additional spacy models

A number of notebooks also use spacy.

The library is already installed when you install wetsuite itself, 
but you can add any number of specific language models. 

None are installed out of the box, and our examples and code currently uses the following four. 

To avoid us needing to add additional install instructions everywhere,
running the four _now_ should cover all the models we use in code and examples:

```python
!python -m spacy download nl_core_news_lg
```

```python
!python -m spacy download en_core_web_lg        
```

```python
!python -m spacy download en_core_web_trf    
```

```python
!python -m spacy download xx_sent_ud_sm
```

  Those four lines, incidentally, also cover most of the spacy models we use in examples.
  
  If we use others, we'll mention additional download lines like this.
